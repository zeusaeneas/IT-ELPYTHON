```python
# Parent class
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        return "Some generic animal sound"

    def info(self):
        return f"{self.name} is a {self.species}."

# Child class 1
class Dog(Animal):
    def make_sound(self):
        return "Bark!"

# Child class 2
class Cat(Animal):
    def make_sound(self):
        return "Meow!"

# Child class 3
class Bird(Animal):
    def make_sound(self):
        return "Tweet!"

# --- Test the classes ---
dog = Dog("Buddy", "Dog")
cat = Cat("Whiskers", "Cat")
bird = Bird("Tweety", "Bird")

print(dog.info(), dog.make_sound())
print(cat.info(), cat.make_sound())
print(bird.info(), bird.make_sound())
```

```python
# Parent Class
class Animal:
    def __init__(self, name, species, habitat):
        self.name = name
        self.species = species
        self.habitat = habitat

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")

# Child Class 1
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Dog", "House")
        self.breed = breed

    def bark(self):
        print(f"{self.name} says Woof!")

# Child Class 2
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, "Cat", "House")
        self.color = color

    def meow(self):
        print(f"{self.name} says Meow!")

# Child Class 3
class Bird(Animal):
    def __init__(self, name, wing_span):
        super().__init__(name, "Bird", "Sky")
        self.wing_span = wing_span

    def fly(self):
        print(f"{self.name} is flying with a wingspan of {self.wing_span} cm.")

# --- Test the classes ---
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers", "White")
bird = Bird("Tweety", 25)

dog.eat()
dog.bark()

cat.sleep()
cat.meow()

bird.eat()
bird.fly()
```

```python
# Parent Class
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")

# Child Class 1
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Dog")
        self.breed = breed

    def show_details(self):
        print(f"{self.name} is a {self.breed} {self.species}.")

    def bark(self):
        print(f"{self.name} says Woof!")

# Child Class 2
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, "Cat")
        self.color = color

    def show_details(self):
        print(f"{self.name} is a {self.color} {self.species}.")

    def meow(self):
        print(f"{self.name} says Meow!")

# Child Class 3
class Bird(Animal):
    def __init__(self, name, wing_span):
        super().__init__(name, "Bird")
        self.wing_span = wing_span

    def show_details(self):
        print(f"{self.name} is a {self.species} with a wingspan of {self.wing_span} cm.")

    def fly(self):
        print(f"{self.name} is flying high in the sky!")

# --- Test the classes ---
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers", "white")
bird = Bird("Tweety", 25)

# Show details and behaviors
dog.show_details()
dog.bark()
dog.eat()
dog.sleep()

print()

cat.show_details()
cat.meow()
cat.eat()
cat.sleep()

print()

bird.show_details()
bird.fly()
bird.eat()
bird.sleep()
```
