# Q10. Robust Temperature Converter (Functions + Exceptions).

## Prompt (paste verbatim)

Write two functions: cel_to_far(c: float) and far_to_cel(f: float) with docstrings and type hints. 
Build a CLI to choose conversion direction, validate input, handle errors, and include unit tests for typical and edge cases like -40.


## ChatGPT Code (paste verbatim)

```python
def cel_to_far(c: float) -> float:
    """Convert Celsius to Fahrenheit."""
    return (c * 9 / 5) + 32

def far_to_cel(f: float) -> float:
    """Convert Fahrenheit to Celsius."""
    return (f - 32) * 5 / 9

def get_float_input(prompt: str) -> float:
    """Safely get a float input from the user."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a number.")

def main():
    """CLI for robust temperature conversion."""
    while True:
        choice = input("Convert (C)elsius to Fahrenheit or (F)ahrenheit to Celsius? (Q to quit): ").strip().upper()
        if choice == "C":
            temp_c = get_float_input("Enter temperature in Celsius: ")
            print(f"{temp_c}°C = {cel_to_far(temp_c):.2f}°F")
        elif choice == "F":
            temp_f = get_float_input("Enter temperature in Fahrenheit: ")
            print(f"{temp_f}°F = {far_to_cel(temp_f):.2f}°C")
        elif choice == "Q":
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please type C, F, or Q.")

if __name__ == "__main__":
    main()
```

## Critique

**CORRECTNESS:** Formulae for Celsius ↔ Fahrenheit are correct.
                 Handles basic input/output flow.

**COMPLEXITY:** O(1) for conversions; constant space. Optimal.

**ROBUSTNESS:** Does not handle non-numeric input; float(input(...)) will raise ValueError.
                Only accepts uppercase choice "C" or "F" (user may type lowercase).
                Edge cases like extremely large numbers or empty input not handled.

**READABILITY:** Function names and docstrings are good.
                 CLI code could be wrapped in a loop for repeated use.
                 No comments for user guidance in code.

**FAITHFULNESS:** Uses type hints, docstrings, asserts for testing, and simple control flow.
                  Needs exception handling and better input validation to align with robust function handling discussed in lecture 3 and 4 topic.
